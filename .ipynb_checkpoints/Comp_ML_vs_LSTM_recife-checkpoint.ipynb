```python
import pickle 
from time import time
from models.lstm import train, evaluate
from models.plots_lstm import * 
from models.upload_data import get_cluster_data, get_city_names
from models.preprocessing import normalize_data, split_data
import tensorflow.keras as  keras
from tensorflow.keras.layers import LSTM
from tensorflow.keras import backend as K
from tensorflow.keras.utils import plot_model
from tensorflow.keras.layers import Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from sklearn.metrics import mean_squared_error as mse 
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_absolute_error as mae
```

```python
def get_dengue_data(city, state, end_date,cluster= False,data_types = ['alerta', 'weather'], 
                    predictors = ['casos', 'casos_est', 'casos_est_min',
                                 'casos_est_max', 'p_rt1', 'p_inc100k', 'temp_min',
                                  'temp_max', 'umid_min', 'pressao_min']
                    ):
    
    with open(f'clusters/clusters_{state}.pkl', 'rb') as fp:
        clusters = pickle.load(fp)
        
    data_dengue, group_dengue = get_cluster_data(city, clusters=clusters, data_types= data_types, cols= predictors, doenca='dengue')   
    
    if cluster == False: 
        for i in data_dengue.columns:
        
            if i.endswith(f'{city}') == False:

                del data_dengue[i]


    target_col = list(data_dengue.columns).index("casos_est_{}".format(city))
    
    data_dengue[f'casos_est_{city}_diff'] = np.concatenate(([np.nan], np.diff(data_dengue[f'casos_est_{city}'], 1)), axis = 0)
    
    data_dengue = data_dengue.dropna()
    
    norm_data_dengue, max_features_t = normalize_data(data_dengue.loc[:end_date])
    factor_dengue = max_features_t[target_col]

    ##split test and train
    X_train, Y_train, X_test, Y_test = split_data(
            norm_data_dengue,
            look_back=4,
            ratio=0.75,
            predict_n=4,
            Y_column=target_col,
    )

    return data_dengue.loc[:end_date],factor_dengue,  X_train, Y_train, X_test, Y_test

def dengue_prediction_model(model, city, state, epochs, label_model):
    
    with open(f'data/dengue_{city}.pkl', 'rb') as fp:
        data = pickle.load(fp)
        
    data_dengue = data['data_dengue']
    factor_dengue = data['factor']
    X_train = data['X_train']
    Y_train = data['Y_train'] 
    X_test = data['X_test']
    Y_test = data['Y_test']

    train(label_model, model, X_train, Y_train, batch_size=1, epochs=epochs, geocode= city, overwrite=True)
    
    pred_in = evaluate(model, X_train, Y_train, uncertainty = True)
    pred_out = evaluate(model, X_test, Y_test, uncertainty = True)
    
    preds = np.concatenate( (pred_in, pred_out), axis = 0)

    df_predicted_d = pd.DataFrame(np.percentile(preds, 50, axis=2))
    df_predicted25_d = pd.DataFrame(np.percentile(preds, 2.5, axis=2))
    df_predicted975_d = pd.DataFrame(np.percentile(preds, 97.5, axis=2))

    indice = list(data_dengue.index)
    indice = [i.date() for i in indice]

    plot_predicted_vs_data(preds, np.concatenate( (Y_train, Y_test), axis = 0 ), indice, city, 4, factor_dengue, split_point=len(Y_train), uncertainty= True, 
                        label_name = f'pred_dengue_{city}_{label_model}')  
    
    df = pd.DataFrame(columns = ['model', 'city', 'rmse_in', 'rmse_out'] )
    df['city'] = [city]
    df['model'] = [label_model]
    df['rmse_in'] = [mse(factor_dengue*np.percentile(pred_in, 50, axis = 2), factor_dengue*Y_train, squared = False) ]
    df['rmse_out'] = [mse(factor_dengue*np.percentile(pred_out, 50, axis = 2), factor_dengue*Y_test, squared = False)]


    return df
```

```python
def model_lstm(hidden, features, predict_n, look_back=10, batch_size=1):
    """
    Builds and returns the LSTM model with the parameters given
    :param hidden: number of hidden nodes
    :param features: number of variables in the example table
    :param look_back: Number of time-steps to look back before predicting
    :param batch_size: batch size for batch training
    :return:
    """
    inp = keras.Input(
        shape=(look_back, features),
        # batch_shape=(batch_size, look_back, features)
    )
    
    x = Bidirectional(LSTM(
        hidden,
        input_shape=(look_back, features),
        stateful=False,
        batch_input_shape=(batch_size, look_back, features),
        return_sequences=True,
        #activation='sigmoid',
        dropout=0.1,
        recurrent_dropout=0.1,
        implementation=2,
        unit_forget_bias=True,
    ), merge_mode = 'ave', name = 'bidirectional_1')(inp, training=True)     

    x = Dropout(0.2, name='dropout_1')(x, training=True)
    
    x = LSTM(
        hidden,
        input_shape=(look_back, features),
        stateful=False,
        batch_input_shape=(batch_size, look_back, features),
        return_sequences=True,
        #activation='sigmoid',
        dropout=0.1,
        recurrent_dropout=0.1,
        implementation=2,
        unit_forget_bias=True, name='lstm_1'
    )(x, training=True)

    x = Dropout(0.2, name = 'dropout_2')(x, training=True)
    
    x = LSTM(
        hidden,
        input_shape=(look_back, features),
        stateful=False,
        batch_input_shape=(batch_size, look_back, features),
        return_sequences=False,
        #activation='sigmoid',
        dropout=0.1,
        recurrent_dropout=0.1,
        implementation=2,
        unit_forget_bias=True, name='lstm_2'
    )(x, training=True)

    x = Dropout(0.2, name = 'dropout_3')(x, training=True)

    out = Dense(
        predict_n,
        activation="relu",
        kernel_initializer="random_uniform",
        bias_initializer="zeros",
        name = 'dense'
    )(x)
    model = keras.Model(inp, out)

    start = time()
    model.compile(loss="msle", optimizer="adam", metrics=["accuracy", "mape", "mse"])
    print("Compilation Time : ", time() - start)
    plot_model(model, to_file="LSTM_model.png")
    print(model.summary())
    return model
```

```python
import getpass, os
import pickle
os.environ['PSQL_USER']='dengueadmin'
os.environ['PSQL_HOST']='localhost'
os.environ['PSQL_DB']='dengue'
os.environ['PSQL_PASSWORD']=getpass.getpass("Enter the database password: ")
```

```python
from models.upload_data import get_alerta_table, get_cluster_data, get_city_names  
from models.preprocessing import build_lagged_features 
from sklearn.model_selection import train_test_split
from pgbm_nb import PGBMRegressor,PGBM
from models.plots_qlgbm import plot_prediction

DATA_TYPES = ['alerta', 'weather']
PREDICTORS =  ['casos', 'casos_est', 'casos_est_min', 'casos_est_max', 'p_rt1', 'p_inc100k', 'temp_min', 'temp_max', 'umid_min', 'pressao_min']


def qf_prediction(city, state, horizon, lookback, doenca = 'dengue', split = 0.75, end_date = '2021-01-01'):
    """
    Train model for a single city.
    :param city:
    :param state:
    :param horizon:
    :param lookback:
    :return:
    """
    with open(f'clusters/clusters_{state}.pkl', 'rb') as fp:
        clusters = pickle.load(fp)
    data, group = get_cluster_data(city, clusters=clusters, data_types=DATA_TYPES, cols=PREDICTORS, doenca=doenca)

    for i in data.columns:

        data[f'diff_{i}'] = np.concatenate( ([np.nan], np.diff(data[i], 1)))

    data = data.dropna()

    target = f'casos_est_{city}'
    casos_est_columns = [f'casos_est_{i}' for i in group]
    # casos_columns = ['casos_{}'.format(i) for i in group]

    # data = data_full.drop(casos_columns, axis=1)
    data_lag = build_lagged_features(data, lookback)
    data_lag.dropna()
    
    data_lag = data_lag.loc[:end_date]

    X_data = data_lag.copy()
    
    X_data.drop(target, inplace=True, axis=1)

    targets = {}
    for d in range(1, horizon + 1):
        targets[d] = data_lag[target].shift(-(d))[:-(d)]
        
    X_train, X_test, y_train, y_test = train_test_split(X_data, data_lag[target],
                                                        train_size=split, test_size=1 - split, shuffle=False)

    city_name = get_city_names([city, 0])[0][1]
    preds = np.empty((len(data_lag), horizon))
    preds25 = np.empty((len(data_lag), horizon))
    preds975 = np.empty((len(data_lag), horizon))

    for d in range(1, horizon + 1):
        tgt = targets[d][:len(X_train)]
        tgtt = targets[d][len(X_train):]
        
        model = PGBMRegressor(n_estimators= 100,  distribution='poisson')
        
        model.fit(X_train, tgt)

        model.save(f'saved_models/{city}_{doenca}_city_model_{d}_pgbm.pt')

        pred = model.predict(X_data[:len(targets[d])].values)
        
        pred_dist = model.predict_dist(X_data[:len(targets[d])].values)
        
        pred25 = pred_dist.max(axis=0)
        pred = pred
        pred975 = pred_dist.min(axis=0)
        dif = len(data_lag) - len(pred)
        if dif > 0:
            pred = list(pred) + ([np.nan] * dif)
            pred25 = list(pred25) + ([np.nan] * dif)
            pred975 = list(pred975) + ([np.nan] * dif)
        preds[:, (d - 1)] = pred
        preds25[:, (d - 1)] = pred25
        preds975[:, (d - 1)] = pred975

    x, y, y25, y975 = plot_prediction(preds, preds25, preds975, data_lag[target], f'Predictions for {city_name}', state, len(X_train), doenca,
                                    label = f'pgbm_{city}')

    with open(f'predictions/pgbm_{city}_{doenca}_predictions.pkl', 'wb') as f:
        pickle.dump({'target':data_lag[target],'dates': x, 'preds': y, 'preds25': y25,
                    'preds975': y975, 'city_name': city_name, 'train_size': len(X_train)
                    }, f)
        
    df = pd.DataFrame(columns = ['model', 'city', 'rmse_in', 'rmse_out'] )
    df['city'] = [city]
    df['model'] = ['pgbm']
    df['rmse_in'] = [ mse( data_lag[target][4 :4 + len(X_train)], y[:len(X_train)], squared = False) ]
    df['rmse_out'] = [mse( data_lag[target][4 + len(X_train):], y[len(X_train):], squared = False)]


    return preds, preds25, preds975, X_train, data_lag[target], data_lag, df
```

```python
%%time
horizon = 4
lookback = 4
preds, preds25, preds975, X_train, target, data_lag,df = qf_prediction(city, state, horizon, lookback, doenca = 'dengue', split = 0.75, end_date = '2022-01-01')
```
