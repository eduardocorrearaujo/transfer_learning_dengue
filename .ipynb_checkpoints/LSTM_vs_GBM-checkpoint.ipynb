```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error as mse
```

```python
def plot_predicted_vs_data_n(df_predicted,df_predicted25, df_predicted975,  Ydata, indice, label, pred_window, factor, split_point=None, uncertainty=False, label_name = 'predict'):
    """
    Plot the model's predictions against data
    :param predicted: model predictions
    :param Ydata: observed data
    :param indice:
    :param label: Name of the locality of the predictions
    :param pred_window:
    :param factor: Normalizing factor for the target variable
    """

    plt.figure(figsize=(8,6))
   
    uncertainty = True
    ymax = max( max(df_predicted975.max()) * factor, Ydata.max() * factor)

    if split_point != None: 
        plt.vlines(indice[split_point], 0, ymax, "g", "dashdot", lw=2, label = 'Train/Test')
        
    #P.text(indice[split_point + 2], 0.6 * ymax, "Out of sample Predictions")
    # plot only the last (furthest) prediction point
    plt.plot(indice[len(indice)-Ydata.shape[0]:], Ydata[:, -1] * factor, 'k-', alpha=0.7, label='data')
    plt.plot(indice[len(indice)-Ydata.shape[0]:], df_predicted.iloc[:,-1] * factor, 'r-', alpha=0.5, label='median')
    if uncertainty:
        plt.fill_between(indice[7:], df_predicted25[df_predicted25.columns[-1]] * factor,
                       df_predicted975[df_predicted975.columns[-1]] * factor,
                       color='b', alpha=0.3)


    tag = '_unc' if uncertainty else ''
    plt.grid()
    #P.title("Predictions for {}".format(label))
    plt.xlabel("time")
    plt.ylabel("incidence")
    plt.xticks(rotation=0)
    plt.legend()
    plt.title('LSTM')
    plt.show()
    
    return 
```

```python
city_name = 'Recife'
data_lstm = pickle.load(open(f'predictions/lstm_{city_name}_dengue_predictions.pkl', 'rb'))

plot_predicted_vs_data_n(data_lstm['pred'], data_lstm['lb'], data_lstm['ub'], data_lstm['target'], data_lstm['indice'], city_name, 4, data_lstm['factor'], split_point=len(data_lstm['xdata']), uncertainty=True, label_name = f'pred_dengue_{city_name.lower()}')
    
```

Computing the LSTM error in the test sample:

```python
factor = data_lstm['factor']
split_point=len(data_lstm['xdata'])
mse(data_lstm['target'][split_point:, -1] * factor, data_lstm['pred'][data_lstm['pred'].columns[-1]][split_point:] * factor, squared = False)
```
