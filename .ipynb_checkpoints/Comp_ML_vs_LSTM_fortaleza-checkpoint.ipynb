```python
import getpass, os
import numpy as np
import pandas as pd
import pickle
os.environ['PSQL_USER']='dengueadmin'
os.environ['PSQL_HOST']='localhost'
os.environ['PSQL_DB']='dengue'
os.environ['PSQL_PASSWORD']=getpass.getpass("Enter the database password: ")
```

```python
from models.upload_data import get_alerta_table, get_cluster_data, get_city_names  
from models.preprocessing import build_lagged_features 
from sklearn.model_selection import train_test_split
from pgbm_nb import PGBMRegressor,PGBM
from models.plots_qlgbm import plot_prediction
from sklearn.metrics import mean_squared_error as mse 
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_absolute_error as mae

DATA_TYPES = ['alerta', 'weather']
PREDICTORS =  ['casos', 'casos_est', 'casos_est_min', 'casos_est_max', 'p_rt1', 'p_inc100k', 'temp_min', 'temp_max', 'umid_min', 'pressao_min']


def qf_prediction(city, state, horizon, lookback, doenca = 'dengue', split = 0.75, end_date = '2021-01-01'):
    """
    Train model for a single city.
    :param city:
    :param state:
    :param horizon:
    :param lookback:
    :return:
    """
    with open(f'clusters/clusters_{state}.pkl', 'rb') as fp:
        clusters = pickle.load(fp)
    data, group = get_cluster_data(city, clusters=clusters, data_types=DATA_TYPES, cols=PREDICTORS, doenca=doenca)

    for i in data.columns:

        data[f'diff_{i}'] = np.concatenate( ([np.nan], np.diff(data[i], 1)))

    data = data.dropna()

    target = f'casos_est_{city}'
    casos_est_columns = [f'casos_est_{i}' for i in group]
    # casos_columns = ['casos_{}'.format(i) for i in group]

    # data = data_full.drop(casos_columns, axis=1)
    data_lag = build_lagged_features(data, lookback)
    data_lag.dropna()
    
    data_lag = data_lag.loc[:end_date]

    X_data = data_lag.copy()
    
    X_data.drop(target, inplace=True, axis=1)

    targets = {}
    for d in range(1, horizon + 1):
        targets[d] = data_lag[target].shift(-(d))[:-(d)]
        
    X_train, X_test, y_train, y_test = train_test_split(X_data, data_lag[target],
                                                        train_size=split, test_size=1 - split, shuffle=False)

    city_name = get_city_names([city, 0])[0][1]
    preds = np.empty((len(data_lag), horizon))
    preds25 = np.empty((len(data_lag), horizon))
    preds975 = np.empty((len(data_lag), horizon))

    for d in range(1, horizon + 1):
        tgt = targets[d][:len(X_train)]
        tgtt = targets[d][len(X_train):]
        
        model = PGBMRegressor(n_estimators= 100,  distribution='poisson')
        
        model.fit(X_train, tgt)

        model.save(f'saved_models/{city}_{doenca}_city_model_{d}_pgbm.pt')

        pred = model.predict(X_data[:len(targets[d])].values)
        
        pred_dist = model.predict_dist(X_data[:len(targets[d])].values)
        
        pred25 = pred_dist.max(axis=0)
        pred = pred
        pred975 = pred_dist.min(axis=0)
        dif = len(data_lag) - len(pred)
        if dif > 0:
            pred = list(pred) + ([np.nan] * dif)
            pred25 = list(pred25) + ([np.nan] * dif)
            pred975 = list(pred975) + ([np.nan] * dif)
        preds[:, (d - 1)] = pred
        preds25[:, (d - 1)] = pred25
        preds975[:, (d - 1)] = pred975

    x, y, y25, y975 = plot_prediction(preds, preds25, preds975, data_lag[target], f'Predictions for {city_name}', state, len(X_train), doenca,
                                    label = f'pgbm_{city}')

    with open(f'predictions/pgbm_{city}_{doenca}_predictions.pkl', 'wb') as f:
        pickle.dump({'target':data_lag[target],'dates': x, 'preds': y, 'preds25': y25,
                    'preds975': y975, 'city_name': city_name, 'train_size': len(X_train)
                    }, f)
        
    df = pd.DataFrame(columns = ['model', 'city', 'rmse_in', 'rmse_out'] )
    df['city'] = [city]
    df['model'] = ['pgbm']
    df['rmse_in'] = [ mse( data_lag[target][4 :4 + len(X_train)], y[:len(X_train)], squared = False) ]
    df['rmse_out'] = [mse( data_lag[target][4 + len(X_train):], y[len(X_train):], squared = False)]


    return preds, preds25, preds975, X_train, data_lag[target], data_lag, df
```

```python
%%time
city = 2304400
state = 'CE'
horizon = 4
lookback = 4
preds, preds25, preds975, X_train, target, data_lag,df = qf_prediction(city, state, horizon, lookback, doenca = 'dengue', split = 1.00, end_date = '2016-01-01')
```
