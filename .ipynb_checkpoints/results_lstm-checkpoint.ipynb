Comparing the predictions 

```python
city = 2611606
STATE = 'PE'
doenca = 'chik'
with open(f'predictions/lstm_{city}_fine_transf_chik_predictions.pkl', 'rb') as f:
            data_t = pickle.load(f)
        
with open(f'predictions/lstm_{city}_dengue_model_chik_predictions.pkl', 'rb') as f:
            data_c = pickle.load(f)
        
city_name = data_c['city_name']
        
plot_transf_predicted_vs_data(data_t['pred'], data_c['pred'], 
                       data_c['target'], indice = data_c['indice'] ,  label = f'{city_name} - chik', pred_window = 4,factor = data_c['factor'],
                       split_point =  len(data_c['xdata']), uncertainty = True, 
                      label_name = f'transf_{city}_chik', label1 = 'dengue model', label2 = 'transf model')
```

```python
pw = 1
city = 2611606
city_name = 'Recife'
STATE = 'PE'
with open(f'predictions/lstm_{city}_fine_transf_chik_predictions.pkl', 'rb') as f:
            data = pickle.load(f)
            predicted = data['pred'][pw].values
            real = data['target'][:, pw]
qp, qo = predicted_vs_observed(predicted, real, city, STATE, doenca = 'chik', model_name = 'lstm', city_name = city_name)
print("Predicted: ", np.median(qp - qo), np.percentile(qp - qo, 25), np.percentile(qp - qo, 75))
```
