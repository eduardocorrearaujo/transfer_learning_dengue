```python
def get_chik_data(city, state, cluster = False, ini_date = '2021-01-01', end_date_train = '2021-03-01',  end_date = '2022-03-31',
                 data_types = ['alerta', 'weather'], 
                predictors = ['casos', 'casos_est', 'casos_est_min',
                                 'casos_est_max', 'p_rt1', 'p_inc100k', 'temp_min',
                                  'temp_max', 'umid_min', 'pressao_min']
                ):
    
    with open(f'clusters/clusters_{state}.pkl', 'rb') as fp:
        clusters = pickle.load(fp)
        
    data_chik, group_chik = get_cluster_data(city, clusters=clusters, data_types=data_types, cols= predictors, doenca='chik')

    if cluster == False:
        for i in data_chik.columns:
    
            if i.endswith(f'{city}') == False:

                del data_chik[i]
    
    target_col = list(data_chik.columns).index("casos_est_{}".format(city))
    
    data_chik[f'casos_est_{city}_diff'] = np.concatenate(([np.nan], np.diff(data_chik[f'casos_est_{city}'], 1)), axis = 0)

    data_chik = data_chik.loc[ini_date:end_date] 

    data_chik = data_chik.dropna() 
    
    norm_data_chik, max_features = normalize_data(data_chik)
    
    factor_chik = max_features[target_col]

    X_train, Y_train, X_test_chik, Y_test_chik = split_data(
            norm_data_chik.loc[:data_chik.loc[:end_date_train].shape[0] -1],
            look_back=4,
            ratio=1,
            predict_n=4,
            Y_column=target_col,
    )

    X_pred, Y_pred, X_test_chik, Y_test_chik = split_data(
            norm_data_chik,
            look_back=4,
            ratio=1,
            predict_n=4,
            Y_column=target_col,
    )

    return data_chik, factor_chik, X_train, Y_train, X_pred, Y_pred
```
