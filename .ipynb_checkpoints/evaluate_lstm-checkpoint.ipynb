```python
def get_dengue_data(city, state, end_date,cluster= False,data_types = ['alerta', 'weather'], 
                    predictors = ['casos', 'casos_est', 'casos_est_min',
                                 'casos_est_max', 'p_rt1', 'p_inc100k', 'temp_min',
                                  'temp_max', 'umid_min', 'pressao_min']
                    ):
    
    with open(f'clusters/clusters_{state}.pkl', 'rb') as fp:
        clusters = pickle.load(fp)
        
    data_dengue, group_dengue = get_cluster_data(city, clusters=clusters, data_types= data_types, cols= predictors, doenca='dengue')   
    
    if cluster == False: 
        for i in data_dengue.columns:
        
            if i.endswith(f'{city}') == False:

                del data_dengue[i]


    target_col = list(data_dengue.columns).index("casos_est_{}".format(city))
    
    data_dengue[f'casos_est_{city}_diff'] = np.concatenate(([np.nan], np.diff(data_dengue[f'casos_est_{city}'], 1)), axis = 0)
    
    data_dengue = data_dengue.dropna()
    
    norm_data_dengue, max_features_t = normalize_data(data_dengue.loc[:end_date])
    factor_dengue = max_features_t[target_col]

    ##split test and train
    X_train, Y_train, X_test, Y_test = split_data(
            norm_data_dengue,
            look_back=4,
            ratio=0.75,
            predict_n=4,
            Y_column=target_col,
    )

    return data_dengue.loc[:end_date],factor_dengue,  X_train, Y_train, X_test, Y_test
```

```python
def model_1(hidden, features, predict_n, look_back=10, batch_size=1):
    """
    Builds and returns the LSTM model with the parameters given
    :param hidden: number of hidden nodes
    :param features: number of variables in the example table
    :param look_back: Number of time-steps to look back before predicting
    :param batch_size: batch size for batch training
    :return:
    """
    inp = keras.Input(
        shape=(look_back, features),
        # batch_shape=(batch_size, look_back, features)
    )
    
    x = LSTM(
        hidden,
        input_shape=(look_back, features),
        stateful=False,
        batch_input_shape=(batch_size, look_back, features),
        return_sequences= False,
        # activation='relu',
        dropout=0.1,
        recurrent_dropout=0.1,
        implementation=2,
        unit_forget_bias=True, name='lstm_1'
    )(inp, training=True)

    x = Dropout(0.2, name = 'dropout_1')(x, training=True)
    
    out = Dense(
        predict_n,
        activation="relu",
        kernel_initializer="random_uniform",
        bias_initializer="zeros",
        name = 'dense'
    )(x)
    model = keras.Model(inp, out)

    start = time()
    model.compile(loss="msle", optimizer="adam", metrics=["accuracy", "mape", "mse"])
    print("Compilation Time : ", time() - start)
    plot_model(model, to_file="LSTM_model.png")
    print(model.summary())
    return model
```

```python
def dengue_prediction_model(model, city, state, epochs, label_model):
    
    with open(f'data/dengue_{city}.pkl', 'rb') as fp:
        data = pickle.load(fp)
        
    data_dengue = data['data_dengue']
    factor_dengue = data['factor']
    X_train = data['X_train']
    Y_train = data['Y_train'] 
    X_test = data['X_test']
    Y_test = data['Y_test']

    train('dengue', model, X_train, Y_train, batch_size=1, epochs=epochs, geocode= city, overwrite=True)
    
    pred_in = evaluate(model, X_train, Y_train, uncertainty = True)
    pred_out = evaluate(model, X_test, Y_test, uncertainty = True)
    
    preds = np.concatenate( (pred_in, pred_out), axis = 0)

    df_predicted_d = pd.DataFrame(np.percentile(preds, 50, axis=2))
    df_predicted25_d = pd.DataFrame(np.percentile(preds, 2.5, axis=2))
    df_predicted975_d = pd.DataFrame(np.percentile(preds, 97.5, axis=2))

    indice = list(data_dengue.index)
    indice = [i.date() for i in indice]

    plot_predicted_vs_data(preds, np.concatenate( (Y_train, Y_test), axis = 0 ), indice, city, 4, factor_dengue, split_point=len(Y_train), uncertainty= True, 
                        label_name = f'pred_dengue_{city}_{label_model}')  
    
    df = pd.DataFrame(columns = ['model', 'city', 'mae_in', 'mae_out'] )
    df['city'] = [city]
    df['model'] = [label_model]
    df['mae_in'] = [mae(factor_dengue*np.percentile(pred_in, 50, axis = 2), factor_dengue*Y_train)]
    df['mae_out'] = [mae(factor_dengue*np.percentile(pred_out, 50, axis = 2), factor_dengue*Y_test)]


    return df
```
